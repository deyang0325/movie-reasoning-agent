```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import SentenceTransformer
import faiss
import pickle
import numpy as np

# Google Drive 마운트
from google.colab import drive
drive.mount('/content/drive')

# 인덱스 및 메타데이터 로드
index = faiss.read_index("/content/drive/MyDrive/Colab Notebooks/movie_index.faiss")
with open("/content/drive/MyDrive/Colab Notebooks/movie_sentences.pkl", "rb") as f:
    sentences = pickle.load(f)
with open("/content/drive/MyDrive/Colab Notebooks/movie_metas.pkl", "rb") as f:
    metas = pickle.load(f)

# 한국어 특화 SentenceTransformer 모델 (SNUNLP SBERT) 사용
sentence_model = SentenceTransformer("snunlp/KR-SBERT-V40K-klueNLI-augSTS", device='cuda')

# LLaMA3 모델 로드
model_id = "NousResearch/Meta-Llama-3-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", torch_dtype="auto")
llama3 = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
# 키워드 추출 함수
import json
import re

# 개선된 키워드 추출 함수 (Function Calling 스타일)
def extract_keywords_llama3(text, model_pipe):
    prompt = f"""[INST] 당신은 사용자가 기억하는 영화 설명으로부터 핵심 키워드를 추출하는 전문가입니다.
다음 문장에서 중요한 단어 또는 구절을 **한국어로** 3~5개 추출하세요.

영화 설명:
\"{text}\"

출력은 다음 JSON 형식으로 반환하세요 (단어는 한국어):
{{ "keywords": ["단어1", "단어2", "단어3", "단어4", "단어5"] }}
[/INST]"""

    try:
        output = model_pipe(prompt, max_new_tokens=100)[0]['generated_text']
        # JSON 부분만 추출
        json_str = re.search(r'\{.*\}', output, re.DOTALL)
        if json_str:
            data = json.loads(json_str.group())
            return data.get("keywords", [])
        else:
            raise ValueError("JSON 형식이 아님")
    except Exception as e:
        print("키워드 추출 실패:", e)
        return []
```

```python
# 요약 생성 함수
def generate_summary_with_llama3(movie_info_list, model_pipe):
    summary_input = "다음은 검색된 영화 리스트입니다:\n"
    for idx, info in enumerate(movie_info_list):
        summary_input += f"{idx+1}. 제목: {info.get('title', '')}, 장르: {info.get('genre', '')}, 연도: {info.get('year', '')}\n"
    summary_input += "\n이 영화들의 공통점이나 특징을 요약해 주세요. 간단히 3~4줄."

    try:
        result = model_pipe(summary_input, max_new_tokens=150)[0]['generated_text']
        return result.strip()
    except Exception as e:
        print("요약 실패:", e)
        return ""
```

```python
# 검색 결과 출력 함수 (포맷 정리)
def print_search_results(results, distances):
    print("\n검색 결과 (상위 3개):")
    for rank, (meta, score) in enumerate(zip(results, distances), start=1):
        print(f"{rank}. 제목: {meta.get('title', '[제목 없음]')}")
        print(f"   연도: {meta.get('year', '[연도 없음]')}")
        print(f"   감독: {meta.get('director', '[감독 없음]')}")
        print(f"   주연: {meta.get('actors', '[주연 없음]')}")
        print(f"   줄거리: {meta.get('plot', '[줄거리 없음]')}")
        print(f"   유사도 거리: {score:.4f}\n")
```

```python
# 자동 판별 통합 검색 함수
def run_auto_search():
    user_input = input("기억나는 영화 내용을 입력하세요 (문장 또는 키워드): ").strip()

    is_keyword_mode = "," in user_input or (len(user_input.split()) <= 5 and all(len(w) <= 6 for w in user_input.split()))

    if is_keyword_mode:
        keywords = [kw.strip() for kw in user_input.split(",") if kw.strip()]
        keyword_query = " ".join(keywords)
    else:
        keywords = extract_keywords_llama3(user_input, llama3)
        keyword_query = " ".join(keywords)

    vec = sentence_model.encode([keyword_query]).astype('float32')
    D, I = index.search(vec, k=10)

    results = []
    distances = []
    for score, i in sorted(zip(D[0], I[0]), key=lambda x: x[0])[:3]:
        meta = metas[i]
        if not meta.get("plot"):
            continue
        results.append(meta)
        distances.append(score)

    if results:
        print_search_results(results, distances)
        print("\n요약:")
        print(generate_summary_with_llama3(results, llama3))
    else:
        print("조건에 맞는 영화가 없습니다.")
```

```python
# 실행
import ipywidgets as widgets
from IPython.display import display, clear_output

def run_auto_search_widget():
    input_box = widgets.Text(
        value='',
        placeholder='예: 그 영화 뭐였더라? 기억나는대로 자유롭게 입력해주면 알려드릴게요',
        description='입력:',
        style={'description_width': '50px'},
        layout=widgets.Layout(width='80%')
    )

    search_button = widgets.Button(
        description='검색',
        button_style='info',
        tooltip='입력한 설명 또는 키워드로 영화 검색',
        icon='search'
    )

    output_area = widgets.Output()

    def on_search_clicked(b):
        with output_area:
            clear_output()
            user_input = input_box.value.strip()

            if not user_input:
                print("입력 내용이 없습니다.")
                return

            # 키워드 모드 판단
            is_keyword_mode = "," in user_input or (len(user_input.split()) <= 5 and all(len(w) <= 6 for w in user_input.split()))

            if is_keyword_mode:
                keywords = [kw.strip() for kw in user_input.split(",") if kw.strip()]
                keyword_query = " ".join(keywords)
            else:
                keywords = extract_keywords_llama3(user_input, llama3)
                keyword_query = " ".join(keywords)

            # 문장 임베딩 후 검색
            vec = sentence_model.encode([keyword_query]).astype('float32')
            D, I = index.search(vec, k=10)

            results = []
            distances = []
            for score, i in sorted(zip(D[0], I[0]), key=lambda x: x[0])[:3]:
                meta = metas[i]
                if not meta.get("plot"):
                    continue
                results.append(meta)
                distances.append(score)

            if results:
                print_search_results(results, distances)
                print("\n요약:")
                print(generate_summary_with_llama3(results, llama3))
            else:
                print("조건에 맞는 영화가 없습니다.")

    search_button.on_click(on_search_clicked)

    display(widgets.VBox([input_box, search_button, output_area]))
```

```python
# 기억나는 영화를 자유롭게 입력해주세요
run_auto_search_widget()
```
